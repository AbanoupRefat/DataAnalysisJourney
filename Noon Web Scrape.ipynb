```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time

driver = webdriver.Chrome()
path = 'https://www.noon.com/egypt-en/'
driver.get(path)
print(driver.title)
time.sleep(2)

search_bar = driver.find_element(By.ID, "searchBar")
search_bar.send_keys("samsung")
search_bar.send_keys(Keys.RETURN)
time.sleep(5)

search_results = []

def scrape_current_page():
    product_titles = driver.find_elements(By.CLASS_NAME, "sc-ced09a79-24")
    product_prices = driver.find_elements(By.CLASS_NAME, "amount")
    product_ratings = driver.find_elements(By.CLASS_NAME, "sc-363ddf4f-2")

    for i in range(len(product_titles)):
        product_name = product_titles[i].text
        product_price = product_prices[i].text
    
        try:
            product_rating = product_ratings[i].text
        except IndexError:
            product_rating = "NO RATING"
        
        search_results.append((product_name, product_price, product_rating))

scrape_current_page()

for page in range(4):
    try:
        next_page_button = driver.find_element(By.XPATH, "//li[@class='next']/a[@aria-label='Next page']")
        next_page_button.click()
        time.sleep(5)
        scrape_current_page()
    except Exception as e:
        print(f"Could not navigate to the next page: {e}")
        break

print("Search Results:")
for result in search_results:
    print(result)

driver.close()
```

```python
import pandas as pd

NoonPrice_list = pd.DataFrame(search_results, columns=['Title', 'Price','Rating'])
NoonPrice_list.to_csv('NoonPrice_list.csv')
NoonPrice_list.to_excel('NoonPrice_list.xlsx')
```
