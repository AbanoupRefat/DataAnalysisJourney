```python
import pandas as pd
import numpy as np  
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.common.keys import Keys
import time
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
```

```python
driver = webdriver.Chrome()
path = 'https://www.noon.com/egypt-en/'
driver.get(path)
print(driver.title)
time.sleep(2)
search_bar = driver.find_element(By.ID, "searchBar")
search_bar.send_keys("samsung")
search_bar.send_keys(Keys.RETURN)
time.sleep(5)

search_results = []

def scrape_current_page():
    product_titles = driver.find_elements(By.CLASS_NAME, "sc-26c8c6bb-24")
    product_prices = driver.find_elements(By.CLASS_NAME, "amount")
    product_ratings = driver.find_elements(By.CLASS_NAME, "sc-363ddf4f-2") 

    print(f"Found {len(product_titles)} products on this page.")

    for i in range(len(product_titles)):
        product_name = product_titles[i].text
        product_price = product_prices[i].text
    
        try:
            product_rating = product_ratings[i].text
        except IndexError:
            product_rating = "NO RATING"
        
        print(f"Product: {product_name}, Price: {product_price}, Rating: {product_rating}")
        search_results.append((product_name, product_price, product_rating))


scrape_current_page()

for page in range(4):
    try:
        # Updated XPath for the next page button
        next_page_button = driver.find_element(By.XPATH, '//a[contains(@aria-label, "Next page")]')
        next_page_button.click()
        time.sleep(5)
        scrape_current_page()
    except Exception as e:
        print(f"Could not navigate to the next page: {e}")
        break

print("Search Results:")
for result in search_results:
    print(result)

driver.close()
```

```python
NoonPrice_list = pd.DataFrame(search_results, columns=['Title', 'Price','Rating'])
NoonPrice_list.to_csv('NoonPricelist.csv')
```
