```python
import re
def standardize_score(score):
    # Remove non-numeric characters except colon and handle AET
    score = re.sub(r'[^\d:]', '', score)
    
    # Handle cases like '2:05' to '2:5'
    if re.match(r'^\d+:\d{2}$', score):
        score = score[:-1]
    
    # Ensure the score is in M:N format
    if re.match(r'^\d+:\d+$', score):
        return score
    return 'Invalid Format'

data['Result'] = data['Result'].apply(standardize_score)
```

```python
def split_time(time_str):
    # Match format '90+2' or '45+1'
    match = re.match(r'(\d+)\+(\d+)', time_str)
    if match:
        main_time = int(match.group(1))
        added_time = int(match.group(2))
    else:
        # If no additional time, set Added_Time to 0
        main_time = int(time_str)
        added_time = 0
    return pd.Series([main_time, added_time], index=['Minute', 'Added_Time'])

# Apply the function to the DataFrame
data[['Minute', 'Added_Time']] = data['Minute'].apply(split_time)
```

```python

# Count occurrences of each opponent
opponent_counts = data['Opponent'].value_counts()

# Get the top 10 opponents (or fewer if less than 10 unique opponents)
top5_opponents = opponent_counts.head(5).index
top5_counts = opponent_counts.loc[top5_opponents]

# Filter the original data to include only top 10 opponents
filtered_data = data[data['Opponent'].isin(top5_opponents)]

# Group by opponent and home/away to count occurrences
goal_counts = filtered_data.groupby(['Opponent', 'Home/Away']).size().unstack(fill_value=0)

# Create a pie chart for each opponent
fig, axes = plt.subplots(nrows=1, ncols=5, figsize=(20, 10))
axes = axes.flatten()

for idx, opponent in enumerate(top5_opponents):
    home_goals = goal_counts.loc[opponent, 'Home']
    away_goals = goal_counts.loc[opponent, 'Away']
    sizes = [home_goals, away_goals]
    labels = ['Home Goals', 'Away Goals']
    
    axes[idx].pie(
        sizes,
        labels=labels,
        autopct='%1.1f%%',
        startangle=90,
        colors=['#0033A0','#A4002D']
    )
    axes[idx].set_title(opponent)

plt.tight_layout()
plt.show()
```

```python
# Filter for UEFA Champions League
cl_data = data[data['Competition'] == 'UEFA Champions League']

# Count the number of goals per stage
goals_by_stage = cl_data['Matchday'].value_counts().sort_index()

# Define the stage order for plotting
stage_order = ['Group Stage', 'last 16', 'Quarter-Finals', 'Semi-Finals', 'Final']
goals_by_stage = goals_by_stage.reindex(stage_order, fill_value=0)

# Plotting
plt.figure(figsize=(10, 6))
goals_by_stage.plot(kind='bar', color='royalblue')
plt.title('Goals Scored by Stage in UEFA Champions League')
plt.xlabel('Stage')
plt.ylabel('Number of Goals')
plt.xticks(rotation=45)
plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```
